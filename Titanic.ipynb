```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(["deck", "embark_town", "alive", "who", "adult_male", "class"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
# we will fill the null values of age and embarked
```

```python
df["age"].fillna(df["age"].mean(), inplace = True)
df.info()
```

```python
# CODE TO DROP A WHOLE ROW
```

```python
df.dropna(subset = ["embarked"], inplace = True)
```

```python
df.info()
```

```python
# Now we have to predict survived or not, we will level encode
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
X = df.drop("survived", axis = 1)
y = df["survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Now we use logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

# KNN (k nearest neighbour)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors= 5)
knn_model.fit(X_train,y_train)
```

```python
y_pred_knn = knn_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_knn)
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
print(classification_report(y_test,y_pred_knn))
```

# SUPPORT VECTOR MACHINE (SVM)

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
model_SVC = SVC(kernel = 'rbf')
model_SVC.fit(X_train,y_train)
```

```python
y_pred_SVC = model_SVC.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_SVC)
```

```python
# non scaled performed average
```

```python
model_SVC2 = SVC(kernel = 'rbf')
model_SVC2.fit(X_train_scaled,y_train)
```

```python
y_pred_SVC2 = model_SVC2.predict(X_test_scaled)
```

```python
accuracy_score(y_test,y_pred_SVC2)
```

```python
# hence we can see that svm gives the best accuracy among all of these models using scaled dataset
```

```python
confusion_matrix(y_test,y_pred_SVC2)
```

```python
classification_report(y_test,y_pred_SVC2)
```
